```python
# c9 — Références couleur & assignation persistantes (indices utilisateur, ROI torse, LAB)
import numpy as np
import cv2, os, json

# ========= Pré-requis =========
# Assure-toi d’avoir exécuté c6/c7 pour créer `crops` (liste de crops RGB)
try:
    _ = len(crops)
except NameError:
    raise RuntimeError("❗️Définis d’abord `crops` (via c6/c7) avant c9.")

# ========= Indices CHOISIS PAR L’UTILISATEUR =========
team_A_indices = [0, 1, 3, 4]     # joueurs A
team_B_indices = [2, 5, 7, 8]     # joueurs B
referee_indices = [20, 22, 23]    # arbitres
team_A_GK = [13]                  # GK A (noir)
team_B_GK = []                    # GK B

# ========= Labels cohérents avec la pipeline vidéo =========
TEAM_LABELS = {
    'team_A':    0,
    'team_B':    1,
    'referee':   2,
    'team_A_GK': 3,
    'team_B_GK': 4
}

# ========= Espace couleur =========
USE_AB_ONLY = False
def _to_ab(vec3):
    if vec3 is None: return None
    return vec3[1:] if USE_AB_ONLY else vec3

def extract_jersey_lab(
    crop_rgb,
    torso_margin=(0.25, 0.70),
    side_margin=0.25,
    mask_grass=True,
    min_pixels=120):
    """Extrait la moyenne LAB du torse, avec masque pelouse/ternes optionnels."""
    if crop_rgb is None or crop_rgb.size == 0:
        return None
    h, w = crop_rgb.shape[:2]
    y1 = int(h * torso_margin[0]); y2 = int(h * torso_margin[1])
    x1 = int(w * side_margin);     x2 = int(w * (1 - side_margin))
    if x2 <= x1 or y2 <= y1: return None

    roi = crop_rgb[y1:y2, x1:x2].copy()
    if roi.size == 0: return None

    hsv = cv2.cvtColor(roi, cv2.COLOR_RGB2HSV)
    H, S, V = cv2.split(hsv)
    keep = np.ones(roi.shape[:2], dtype=bool)
    if mask_grass:
        grass = (H >= 35) & (H <= 90) & (S > 55) & (V > 40)
        keep &= ~grass
    low_sv = (S < 35) | (V < 25)
    keep &= ~low_sv
    if keep.sum() < min_pixels:
        keep = np.ones_like(keep)

    lab = cv2.cvtColor(roi, cv2.COLOR_RGB2LAB).reshape(-1, 3)
    lab = lab[keep.reshape(-1)]
    if lab.size == 0: return None
    return np.mean(lab, axis=0)

def _safe_mean_from_indices(indices, *, mask_grass=True):
    vecs = []
    for i in indices:
        if 0 <= i < len(crops):
            v = extract_jersey_lab(crops[i], mask_grass=mask_grass)
            if v is not None:
                vecs.append(_to_ab(v))
    if len(vecs) == 0: return None, 0
    arr = np.stack(vecs, axis=0)
    return np.mean(arr, axis=0), arr.shape[0]

# ========= Références en mémoire =========
team_refs = {
    'team_A':    None,
    'team_B':    None,
    'referee':   None,
    'team_A_GK': None,
    'team_B_GK': None
}

# ========= Pipeline C9 (sans JSON) =========
def append_refs_from_indices(index_dict, counts):
    for key, idxs in index_dict.items():
        if not idxs: continue
        is_gk = key in ("team_A_GK", "team_B_GK")
        mean_new, m = _safe_mean_from_indices(idxs, mask_grass=(not is_gk))
        if mean_new is None or m == 0:
            continue
        old = team_refs.get(key)
        n = counts.get(key, 0)
        if old is None or n == 0:
            team_refs[key] = mean_new
            counts[key] = m
        else:
            team_refs[key] = (old*n + mean_new*m) / (n + m)
            counts[key] = n + m
    print("✅ Références mises à jour (append).")
    return counts

# === Version simplifiée : on ignore totalement le JSON ===
counts = {k: 0 for k in team_refs.keys()}
counts = append_refs_from_indices({
    "team_A":    team_A_indices,
    "team_B":    team_B_indices,
    "referee":   referee_indices,
    "team_A_GK": team_A_GK,
    "team_B_GK": team_B_GK
}, counts)

print("✅ Références calculées uniquement à partir de cette frame (aucune persistance JSON).")

# ========= Log final =========
print("✅ Références (LAB {}):".format("(a,b)" if USE_AB_ONLY else "(L,a,b)"))
for team, color in team_refs.items():
    print(f"  {team:11s}: {color if color is not None else 'None (ignoré)'}")
```

```python
# c13 — Pipeline vidéo : A/B strict (c9) + arbitre via ref c9 (anti-flip joueurs + off-pitch)
#          + GK via GK_refs + ballon robuste
import supervision as sv
import numpy as np
import cv2, os
from collections import deque, Counter, defaultdict

# ==== IDs classes (détecteur) ====
BALL_ID       = 0
GOALKEEPER_ID = 1
PLAYER_ID     = 2
REFEREE_ID    = 3

# ==== Annotateurs (A, B, arbitre) ====
try:
    ellipse_annotator
    label_annotator
    triangle_annotator
except NameError:
    ellipse_annotator = sv.EllipseAnnotator(
        color=sv.ColorPalette.from_hex(["#5F9EF0", "#C24154", "#FFFF00"]),
        thickness=2
    )
    label_annotator = sv.LabelAnnotator(
        color=sv.ColorPalette.from_hex(["#5F9EF0", "#C24154", "#FFFF00"]),
        text_color=sv.Color.from_hex('#000000'),
        text_position=sv.Position.BOTTOM_CENTER
    )
    triangle_annotator = sv.TriangleAnnotator(
        color=sv.Color.from_hex('#FFD700'),
        base=25, height=21, outline_thickness=1
    )

# ==== TEAM_LABELS ====
if 'TEAM_LABELS' not in globals():
    TEAM_LABELS = {'team_A':0,'team_B':1,'referee':2,'team_A_GK':3,'team_B_GK':4}

# ==== Chargement références couleur (depuis C9) ====
try:
    team_refs
except NameError:
    raise RuntimeError("❗️Exécute d’abord la cellule C9 avant C13 (team_refs non défini).")

print("✅ Références couleur chargées depuis C9 (aucune lecture JSON).")

USE_AB_ONLY = False

# ==== Fonctions utilitaires ====
def _to_ab(vec3):
    if vec3 is None: return None
    return vec3[1:] if USE_AB_ONLY else vec3

def extract_jersey_lab(crop_rgb, torso_margin=(0.25,0.70), side_margin=0.25, mask_grass=True, min_pixels=120):
    if crop_rgb is None or crop_rgb.size == 0: return None
    h, w = crop_rgb.shape[:2]
    y1 = int(h*torso_margin[0]); y2 = int(h*torso_margin[1])
    x1 = int(w*side_margin);     x2 = int(w*(1-side_margin))
    if x2<=x1 or y2<=y1: return None
    roi = crop_rgb[y1:y2, x1:x2].copy()
    if roi.size == 0: return None
    hsv = cv2.cvtColor(roi, cv2.COLOR_RGB2HSV); H,S,V = cv2.split(hsv)
    keep = np.ones(roi.shape[:2], dtype=bool)
    if mask_grass:
        grass = (H>=35)&(H<=90)&(S>55)&(V>40); keep &= ~grass
    low_sv = (S<35)|(V<25); keep &= ~low_sv
    if keep.sum() < min_pixels: keep = np.ones_like(keep)
    lab = cv2.cvtColor(roi, cv2.COLOR_RGB2LAB).reshape(-1,3); lab = lab[keep.reshape(-1)]
    if lab.size == 0: return None
    return np.mean(lab, axis=0)

def assign_team_with_scores(crop_rgb, refs, *, mask_grass=True):
    col = extract_jersey_lab(crop_rgb, mask_grass=mask_grass)
    if col is None:
        for k in ['referee','team_A','team_B','team_A_GK','team_B_GK']:
            if refs.get(k) is not None: return TEAM_LABELS[k], {k:0.0}
        return TEAM_LABELS['team_A'], {}
    col = _to_ab(col)
    dists={}
    for name, ref in refs.items():
        if ref is None: continue
        dists[name] = float(np.linalg.norm(col - ref))
    if not dists: return TEAM_LABELS['team_A'], {}
    assigned = min(dists, key=dists.get)
    return TEAM_LABELS[assigned], dists

def assign_team(crop_rgb, refs, *, mask_grass=True):
    lab, _ = assign_team_with_scores(crop_rgb, refs, mask_grass=mask_grass)
    return lab

# ==== Fonctions pour terrain ====
def build_pitch_mask_fast(frame_rgb, h_low=35, h_high=90, s_min=40, v_min=40, morph_ks=7, scale=0.33):
    H, W = frame_rgb.shape[:2]
    small = cv2.resize(frame_rgb, (int(W*scale), int(H*scale)), interpolation=cv2.INTER_AREA)
    hsv   = cv2.cvtColor(small, cv2.COLOR_RGB2HSV)
    Hc, Sc, Vc = cv2.split(hsv)
    grass = (Hc >= h_low) & (Hc <= h_high) & (Sc >= s_min) & (Vc >= v_min)
    mask  = grass.astype(np.uint8) * 255
    if morph_ks > 0:
        k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (morph_ks, morph_ks))
        mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, k)
        mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN,  k)
    num, labels = cv2.connectedComponents(mask)
    if num > 1:
        areas = [(labels==i).sum() for i in range(1, num)]
        biggest = 1 + int(np.argmax(areas))
        mask = (labels == biggest).astype(np.uint8) * 255
    mask_full = cv2.resize(mask, (W, H), interpolation=cv2.INTER_NEAREST)
    return mask_full.astype(bool)

def _is_on_pitch_xyxy(xyxy, pitch_mask_bool, patch_px=12, min_cover=0.40):
    if pitch_mask_bool is None or len(xyxy)==0: return False
    H, W = pitch_mask_bool.shape
    x1,y1,x2,y2 = xyxy.astype(int)
    cx = int(0.5*(x1+x2)); by = int(y2)
    x1p = max(0, cx - patch_px); x2p = min(W, cx + patch_px + 1)
    y1p = max(0, by - patch_px); y2p = min(H, by + patch_px + 1)
    if x2p <= x1p or y2p <= y1p: return False
    patch = pitch_mask_bool[y1p:y2p, x1p:x2p]
    cover = float(patch.sum()) / float(patch.size)
    return cover >= min_cover

def keep_if_on_pitch(dets, pitch_mask_bool, patch_px=8, min_cover=0.25):
    if len(dets.xyxy) == 0: return dets
    H, W = pitch_mask_bool.shape
    keep = np.zeros(len(dets.xyxy), dtype=bool)
    for i, (x1, y1, x2, y2) in enumerate(dets.xyxy.astype(int)):
        cx = int(0.5*(x1 + x2)); by = int(y2)
        x1p = max(0, cx - patch_px); x2p = min(W, cx + patch_px + 1)
        y1p = max(0, by - patch_px); y2p = min(H, by + patch_px + 1)
        if x2p <= x1p or y2p <= y1p:
            keep[i] = False; continue
        patch = pitch_mask_bool[y1p:y2p, x1p:x2p]
        cover = float(patch.sum()) / float(patch.size)
        keep[i] = (cover >= min_cover)
    return dets[keep]

# ==== Tracker ballon robuste ====
_last_ball = {"xyxy": None, "c": None, "v": np.array([0.0, 0.0]), "miss": 0}
BALL_COAST_FRAMES = 20
EMA_VEL_ALPHA = 0.6

def _center(xy):
    x1,y1,x2,y2 = xy
    return np.array([(x1+x2)/2.0, (y1+y2)/2.0], dtype=float)

def track_ball_robust(ball_dets: sv.Detections, frame_rgb) -> sv.Detections:
    global _last_ball
    if len(ball_dets.xyxy) > 0:
        xy = ball_dets.xyxy[0]
        c = _center(xy)
        if _last_ball["c"] is not None:
            v = c - _last_ball["c"]
            _last_ball["v"] = EMA_VEL_ALPHA*v + (1.0-EMA_VEL_ALPHA)*_last_ball["v"]
        _last_ball["xyxy"] = xy
        _last_ball["c"] = c
        _last_ball["miss"] = 0
        return ball_dets
    if _last_ball["xyxy"] is not None and _last_ball["miss"] < BALL_COAST_FRAMES:
        _last_ball["miss"] += 1
        _last_ball["v"] *= 0.75
        c = _last_ball["c"] + _last_ball["v"]
        x1,y1,x2,y2 = _last_ball["xyxy"]
        w = x2-x1; h = y2-y1
        xy = np.array([c[0]-w/2, c[1]-h/2, c[0]+w/2, c[1]+h/2], dtype=float)
        return sv.Detections(xyxy=xy[np.newaxis, :], class_id=np.array([0], dtype=int))
    _last_ball = {"xyxy": None, "c": None, "v": np.array([0.0,0.0]), "miss": 0}
    return sv.Detections.empty()




# ==== Fonctions joueurs / arbitres / GK (identiques à ton code) ====
VOTE_WINDOW   = 5
LOCK_AFTER    = 3
AB_MARGIN_MIN = 10.0
SWITCH_STRONG_VOTES = 6
team_state = {}

def classify_team_stable(frame_rgb, players_dets, frame_idx):
    if len(players_dets.xyxy) == 0:
        return players_dets
    ab_refs = {k: v for k, v in team_refs.items() if k in ['team_A', 'team_B'] and v is not None}
    if len(ab_refs) == 0:
        players_dets.class_id = np.zeros(len(players_dets.xyxy), dtype=int)
        return players_dets
    out = []
    tids = getattr(players_dets, "tracker_id", None)
    for i, xy in enumerate(players_dets.xyxy):
        tid = int(tids[i]) if tids is not None else i
        st = team_state.get(tid)
        if st is None:
            st = {"votes": deque(maxlen=VOTE_WINDOW), "team": None,
                  "first_seen": frame_idx, "strong_opposition": 0}
            team_state[tid] = st
        crop = sv.crop_image(frame_rgb, xy)
        lab, d = assign_team_with_scores(crop, ab_refs)
        vote = int(lab)
        margin = abs(d.get('team_A', 0.0) - d.get('team_B', 0.0)) if ('team_A' in d and 'team_B' in d) else 0.0
        if (st["team"] is not None) and (margin < AB_MARGIN_MIN):
            out.append(st["team"]); continue
        st["votes"].append(vote)
        if st["team"] is None:
            team_guess, n = Counter(st["votes"]).most_common(1)[0]
            if margin >= AB_MARGIN_MIN and n >= LOCK_AFTER:
                st["team"] = int(team_guess)
            out.append(int(st["team"] if st["team"] is not None else team_guess))
            continue
        if vote != st["team"] and margin >= AB_MARGIN_MIN:
            st["strong_opposition"] += 1
            if st["strong_opposition"] >= SWITCH_STRONG_VOTES:
                st["team"] = vote
                st["strong_opposition"] = 0
        else:
            st["strong_opposition"] = 0
        out.append(st["team"])
    players_dets.class_id = np.array(out, dtype=int)
    return players_dets

REF_MARGIN    = 8.0
REF_ABS_MAX   = 12.0
REF_STRONG_VOTES = 5
ref_state = {}
ref_votes = defaultdict(int)

def split_refs_by_color(others_dets, frame_rgb):
    if len(others_dets.xyxy) == 0:
        return sv.Detections.empty(), others_dets
    ref_vec = team_refs.get('referee', None)
    ab_refs = {k: v for k, v in team_refs.items() if k in ['team_A','team_B'] and v is not None}
    if ref_vec is None or len(ab_refs) == 0:
        refs_mask = (others_dets.class_id == REFEREE_ID)
        return others_dets[refs_mask], others_dets[~refs_mask]
    ref_only = {'referee': ref_vec}
    keep_ref = np.zeros(len(others_dets.xyxy), dtype=bool)
    tids = getattr(others_dets, "tracker_id", None)
    for i, xy in enumerate(others_dets.xyxy):
        crop = sv.crop_image(frame_rgb, xy)
        _, dref = assign_team_with_scores(crop, ref_only)
        d_ref = dref.get('referee', 1e9)
        _, dab  = assign_team_with_scores(crop, ab_refs)
        best_ab = min(dab.values()) if len(dab)>0 else 1e9
        is_ref_from_c9 = (d_ref <= REF_ABS_MAX)
        is_ref_like = is_ref_from_c9 or ((best_ab - d_ref) >= REF_MARGIN)
        tid = int(tids[i]) if tids is not None else None
        already_player = (tid is not None and tid in team_state and team_state[tid]["team"] is not None)
        off_pitch = not _is_on_pitch_xyxy(xy, pitch_mask_bool)
        if off_pitch and (is_ref_from_c9 or others_dets.class_id[i] == REFEREE_ID):
            keep_ref[i] = True
            if tid is not None: ref_votes[tid] = max(ref_votes[tid], 1)
            continue
        if is_ref_like and already_player:
            ref_votes[tid] += 1
            keep_ref[i] = (ref_votes[tid] >= REF_STRONG_VOTES)
        elif is_ref_like and not already_player:
            keep_ref[i] = True
            if tid is not None: ref_votes[tid] = max(ref_votes[tid], 1)
        else:
            if tid is not None: ref_votes[tid] = 0
            keep_ref[i] = False
    refs = others_dets[keep_ref]
    rest = others_dets[~keep_ref]
    return refs, rest

def stabilize_refs(ref_dets):
    if len(ref_dets.xyxy) == 0:
        return ref_dets
    tids = getattr(ref_dets, "tracker_id", None)
    if tids is None:
        ref_dets.class_id = np.full(len(ref_dets.xyxy), TEAM_LABELS['referee'], dtype=int)
        return ref_dets
    for tid in tids:
        t = int(tid)
        ref_state[t] = ref_state.get(t, 0) + 1
    ref_dets.class_id = np.full(len(ref_dets.xyxy), TEAM_LABELS['referee'], dtype=int)
    return ref_dets

gk_state = {}

def classify_gk_locked(gk_dets, frame_rgb):
    if len(gk_dets.xyxy) == 0:
        return gk_dets
    tids = getattr(gk_dets, "tracker_id", None)
    out = []
    for i, xy in enumerate(gk_dets.xyxy):
        tid = int(tids[i]) if tids is not None else i
        if tid in gk_state:
            out.append(gk_state[tid]); continue
        crop = sv.crop_image(frame_rgb, xy)
        gk_sub = {k:v for k,v in team_refs.items() if k in ['team_A_GK','team_B_GK'] and v is not None}
        if len(gk_sub) > 0:
            rid = int(assign_team(crop, gk_sub, mask_grass=False))
            final = TEAM_LABELS['team_A'] if rid == TEAM_LABELS['team_A_GK'] else TEAM_LABELS['team_B']
            pal = 0 if final == TEAM_LABELS['team_A'] else 1
        else:
            ab = {k:v for k,v in team_refs.items() if k in ['team_A','team_B'] and v is not None}
            rid = int(assign_team(crop, ab)) if len(ab)>0 else TEAM_LABELS['team_A']
            pal = 0 if rid == TEAM_LABELS['team_A'] else 1
        gk_state[tid] = pal
        out.append(pal)
    gk_dets.class_id = np.array(out, dtype=int)
    return gk_dets

# ==== Writer vidéo ====
vi = sv.VideoInfo.from_video_path(SOURCE_VIDEO_PATH)
W, H, FPS = int(vi.width), int(vi.height), float(vi.fps if vi.fps else 25.0)
writer = cv2.VideoWriter(
    "video_tacticalview_537.mp4",
    cv2.VideoWriter_fourcc(*"mp4v"),
    FPS, (W, H)
)
print(f"✅ Writer: {W}x{H}@{FPS:.2f}fps → video_tacticalview_537.mp4")

frame_gen = sv.get_video_frames_generator(SOURCE_VIDEO_PATH)
frame_idx = 0
pitch_mask_bool = None

# ==== Boucle principale ====
for frame in frame_gen:
    if pitch_mask_bool is None or (frame_idx % 10 == 0):
        pitch_mask_bool = build_pitch_mask_fast(frame)

    result = PLAYER_DETECTION_MODEL.infer(frame, confidence=0.30)[0]
    detections = sv.Detections.from_inference(result)

    ball_raw  = detections[detections.class_id == BALL_ID]
    ball_dets = track_ball_robust(ball_raw, frame)

    others = detections[detections.class_id != BALL_ID]
    others = others.with_nms(threshold=0.50, class_agnostic=True)
    others = tracker.update_with_detections(others)
    others = keep_if_on_pitch(others, pitch_mask_bool)

    refs_raw, non_refs = split_refs_by_color(others, frame)
    gk_raw  = non_refs[non_refs.class_id == GOALKEEPER_ID]
    ppl_raw = non_refs[non_refs.class_id == PLAYER_ID]

    ppl = classify_team_stable(frame, ppl_raw, frame_idx)
    gk = classify_gk_locked(gk_raw, frame)
    refs = stabilize_refs(refs_raw)

    all_dets = sv.Detections.merge([ppl, gk, refs])
    labels = [f"#{tid}" for tid in getattr(all_dets, "tracker_id", [])] \
             if getattr(all_dets, "tracker_id", None) is not None else []

    annotated = frame.copy()
    if len(all_dets.xyxy) > 0:
        annotated = ellipse_annotator.annotate(annotated, all_dets)
        if len(labels) == len(all_dets.xyxy):
            annotated = label_annotator.annotate(annotated, all_dets, labels=labels)
    if len(ball_dets.xyxy) > 0:
        lifted_xy = np.array(ball_dets.xyxy)
        dets_triangle = sv.Detections(xyxy=lifted_xy, class_id=np.zeros(len(lifted_xy), dtype=int))
        annotated = triangle_annotator.annotate(annotated, dets_triangle)

    writer.write(cv2.cvtColor(annotated, cv2.COLOR_RGB2BGR))
    frame_idx += 1

writer.release()
print("✅ Vidéo annotée enregistrée : video_tacticalview_537.mp4")
```

```python
# ===========================================================
# ⚽ POSSESSION (approximation simple)
# ===========================================================
from scipy.spatial.distance import cdist

if "pitch_players_xy" in locals() and len(pitch_players_xy) > 0:
    ball_pos = pitch_ball_xy
    A = pitch_players_xy[ppl.class_id == 0]
    B = pitch_players_xy[ppl.class_id == 1]

    # calcul du joueur le plus proche du ballon à chaque frame
    dist_A = cdist(ball_pos, A).min(axis=1) if len(A)>0 else np.ones(len(ball_pos))*1e6
    dist_B = cdist(ball_pos, B).min(axis=1) if len(B)>0 else np.ones(len(ball_pos))*1e6

    # proportion de frames dominées par chaque équipe
    A_closer = np.sum(dist_A < dist_B)
    B_closer = np.sum(dist_B < dist_A)
    total = A_closer + B_closer
    A_ratio = (A_closer / total * 100) if total>0 else 0
    B_ratio = 100 - A_ratio

    print(f"🕒 Temps de possession estimé : Team A {A_ratio:.1f}% | Team B {B_ratio:.1f}%")
else:
    print("⚠️ Pas de joueurs projetés sur le terrain pour calculer la possession.")
```

```python
# c14.4 — Voronoï de contrôle d’espace (snapshot, robuste aux cas vides)
import numpy as np

def _clean_xy(xy: np.ndarray) -> np.ndarray:
    if xy is None or xy.size == 0:
        return np.empty((0, 2), dtype=float)
    xy = np.asarray(xy, dtype=float)
    # retire NaN/Inf et garde bien (N,2)
    ok = np.isfinite(xy).all(axis=1)
    xy = xy[ok]
    if xy.ndim != 2 or xy.shape[1] != 2:
        return np.empty((0, 2), dtype=float)
    return xy

USE_VORONOI_FALLBACK_SPLIT = True  # si True et qu'une équipe est vide, on split par médiane(X) pour l'affichage

if frame_ref_points.shape[0] >= 4 and len(pitch_players_xy) > 0:
    # Nettoyage
    pitch_players_xy = _clean_xy(pitch_players_xy)
    # Masques classes (si class_id absent plus haut, c14.3 mettait tout en 0 → team_2 vide)
    cls = getattr(ppl, "class_id", None)
    if cls is None or len(cls) != len(pitch_players_xy):
        # sécurité : tous en team 0 si classes manquantes
        cls = np.zeros(len(pitch_players_xy), dtype=int)

    team1_xy = _clean_xy(pitch_players_xy[cls == 0])
    team2_xy = _clean_xy(pitch_players_xy[cls == 1])

    # Fallback si une équipe est vide
    if (team1_xy.shape[0] == 0 or team2_xy.shape[0] == 0) and USE_VORONOI_FALLBACK_SPLIT and pitch_players_xy.shape[0] >= 2:
        xs = pitch_players_xy[:, 0]
        med = np.median(xs)
        left_mask = xs <= med
        # assure au moins 1 point de chaque côté
        if left_mask.sum() == 0 or (~left_mask).sum() == 0:
            # si tout du même côté, on ne force pas ; on skip
            pass
        else:
            team1_xy = _clean_xy(pitch_players_xy[left_mask])
            team2_xy = _clean_xy(pitch_players_xy[~left_mask])

    if team1_xy.shape[0] >= 1 and team2_xy.shape[0] >= 1:
        vor = draw_pitch(CONFIG)
        vor = draw_pitch_voronoi_diagram(
            config=CONFIG,
            team_1_xy=team1_xy,
            team_2_xy=team2_xy,
            team_1_color=sv.Color.from_hex('00BFFF'),
            team_2_color=sv.Color.from_hex('FF1493'),
            pitch=vor
        )
        sv.plot_image(vor)
        print(f"✅ c14.4 Voronoï OK — t1:{len(team1_xy)} pts, t2:{len(team2_xy)} pts")
    else:
        print("⚠️ c14.4 Voronoï sauté : besoin d’au moins 1 joueur projeté par équipe."
              f" (team1={team1_xy.shape[0]}, team2={team2_xy.shape[0]})")
else:
    print("❗️c14.4 Voronoï non dispo (homographie non établie ou aucun joueur projeté).")
```

```python
# ===========================================================
# ⚽ STATISTIQUES BALLE - Distance, vitesse, tirs (robuste)
# ===========================================================
import numpy as np
import matplotlib.pyplot as plt

fps = video_info.fps

# 🔹 Nettoyage du path_raw
valid_points = []
for arr in path_raw:
    if isinstance(arr, np.ndarray) and arr.size > 0:
        arr = np.squeeze(arr)
        if arr.ndim == 1 and arr.shape[0] == 2:
            valid_points.append(arr)
        elif arr.ndim == 2 and arr.shape[1] == 2:
            valid_points.extend(arr)
valid_points = np.array(valid_points, dtype=float)

if valid_points.shape[0] < 2:
    raise ValueError("Pas assez de points valides dans path_raw pour calculer les statistiques.")

# 1️⃣ Distance totale (en unités terrain)
dist = np.linalg.norm(np.diff(valid_points, axis=0), axis=1)
distance_totale = np.sum(dist)
print(f"📏 Distance parcourue par le ballon : {distance_totale:.2f} unités terrain (≈ à calibrer en mètres)")

# 2️⃣ Vitesse instantanée et moyenne
vitesse = dist * fps
vitesse_moy = np.mean(vitesse)
vitesse_max = np.max(vitesse)
print(f"⚡ Vitesse moyenne : {vitesse_moy:.2f} u/s — Max : {vitesse_max:.2f} u/s")

# 3️⃣ Détection de tirs (pics de vitesse)
shoot_threshold = vitesse_moy + 2 * np.std(vitesse)
shoots = np.where(vitesse > shoot_threshold)[0]
print(f"🎯 Tirs détectés (pics de vitesse > seuil) : {len(shoots)}")

# 4️⃣ Graphique de la vitesse
plt.figure(figsize=(10,3))
plt.plot(vitesse, label="Vitesse du ballon (u/s)", color='orange')
plt.axhline(vitesse_moy, color='gray', linestyle='--', label="Moyenne")
for s in shoots:
    plt.axvline(s, color='red', linestyle='--', alpha=0.5)
plt.title("Vitesse instantanée du ballon")
plt.xlabel("Frames"); plt.ylabel("Vitesse (unités/s)")
plt.legend(); plt.show()
```

```python
# ===========================================================
# ⚽ ZONE DE JEU & DIRECTION DU BALLON (corrigée)
# ===========================================================
import numpy as np

# utiliser les points nettoyés calculés plus haut
ball_points = valid_points

x_positions = ball_points[:, 0]
y_positions = ball_points[:, 1]

# ✅ Séparer le terrain en deux moitiés gauche / droite
mid_x = np.median(x_positions)
left_ratio = np.mean(x_positions < mid_x) * 100
right_ratio = 100 - left_ratio

print(f"🧭 Répartition des actions : {left_ratio:.1f}% côté gauche | {right_ratio:.1f}% côté droit")

# ✅ Mesure de progression avant/arrière (variation en X)
forward_moves = np.sum(np.diff(x_positions) > 0)
backward_moves = np.sum(np.diff(x_positions) < 0)

print(f"➡️ Avancées vers l’avant : {forward_moves}")
print(f"⬅️ Retours vers l’arrière : {backward_moves}")
```
